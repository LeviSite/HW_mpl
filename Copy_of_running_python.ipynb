```python
import pandas as pd
import numpy as np
from scipy.stats import t
import seaborn as sns
import matplotlib.pyplot as plt

rv = t(1)
n = 2

vals = []
for i in range(100):
    a = rv.rvs(size = n)
    s = np.mean(a)
    vals.append(s)
    
df = pd.DataFrame({"x": vals})
df["PDF"] = rv.pdf(df["x"])
    
sns.lineplot(data = df, x="x", y="PDF")
sns.histplot (x = vals, stat = "density")
plt.tight_layout()
plt.show()
```
